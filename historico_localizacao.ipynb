```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
import json 

with open ('Histórico de localização.json', encoding='utf8') as file:
    dados = json.load(file)
```

```python
df_historico = pd.json_normalize(dados)
```

```python
df_historico.head()
```

```python
df_historico['timestampMs'] = pd.to_datetime(df_historico['timestampMs'], unit='ms')
```

```python
df_historico.info()
```

```python
df_historico['timestampMs'] = df_historico['timestampMs'].dt.strftime("%Y-%m-%d %H:%M:%S");
```

```python
df_historico['timestampMs'] = pd.to_datetime(df_historico['timestampMs'])
```

```python
df_historico.head()
```

```python
df_historico.info()
```

```python
def atividade(atividade): 
    try: 
        return atividade[0]['activity'][0]['type'] 
    except:
        return 'SEM DADO'
```

```python
def confianca(atividade): 
    try: 
        return atividade[0] ['activity'][0]['confidence'] 
    except:
        return 'SEM DADO'
```

```python
df_historico['atividade'] = df_historico['activity'].apply(atividade)
```

```python
df_historico['confianca'] = df_historico['activity'].apply(confianca)
```

```python
df_historico.drop(columns='activity', inplace=True)
```

```python
df_historico.rename(columns={'timestampMs': 'data_hora', 'latitudeE7': 'latitude', 'longitudeE7': 'longitude',
                    'accuracy': 'precisao', 'verticalAccuracy': 'precisao_altitude',
                    'velocity': 'velocidade'}, inplace=True)
```
