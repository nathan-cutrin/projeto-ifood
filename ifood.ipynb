```python
import pandas as pd
import json
from pathlib import Path
```

```python
p = Path(r'ifood-vamoai.json')
with p.open('r', encoding='utf-8') as f:
    data = json.loads(f.read())
```

```python
df = pd.json_normalize(data)
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
def nome_pedido(pedido):
    try: 
        return pedido[0]['name']
    except:
        return 'SEM DADO'
```

```python
df['nome'] = df['bag.items'].apply(nome_pedido)
```

```python
df.head()
```

```python
df_food = df.loc[:, ['nome', 'shortId', 'createdAt', 'lastStatus', 'details.mode', 'delivery.address.streetName', 
         'delivery.address.city', 'delivery.address.neighborhood',  
         'delivery.address.coordinates.latitude', 'delivery.address.coordinates.longitude',
         'delivery.driver.id', 'delivery.driver.name', 'delivery.driver.modal','merchant.address.city',
         'merchant.address.neighborhood', 'merchant.address.coordinates.latitude', 'merchant.address.coordinates.longitude',
         'merchant.id', 'merchant.name', 'merchant.type', 'payments.total.value', 
         'review.reviewedAt', 'review.score', 'origin.platform']]
```
